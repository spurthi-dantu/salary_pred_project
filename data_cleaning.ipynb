```python
import pandas as pd
```

```python
df = pd.read_csv('data_scraped.csv')

#salary parsing 
df = df[df['Salary Estimate'] != '-1']
```

```python
df['Salary Estimate']
```

```python
df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if '/hr' in x.lower() else 0)
df['monthly'] = df['Salary Estimate'].apply(lambda x: 1 if '/mo' in x.lower() else 0)
```

```python
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
salary_1 = salary.apply(lambda x: x.replace('\n','').replace('$',''))
salary_2 = salary_1.apply(lambda x: x.lower().replace('/yr','').replace('/hr','').replace('/mo',''))
df['Salary Estimate']=salary_2
```

```python
# Define a function to extract the state abbreviation
def extract_state(location):
    split_location = location.split(',')
    if len(split_location) >= 2:
        return split_location[1].strip()  # Strip to remove leading/trailing spaces
    else:
        return location  

# Apply the function to create the 'job_state' column
df['job_state'] = df['Location'].apply(extract_state)
```

```python
df['job_state'] = df['job_state'].replace({'New York State': 'NY', 'California': 'CA'})
```

```python
# Convert 'Founded' column to integers
df['Founded'] = pd.to_numeric(df['Founded'], errors='coerce')  # coerce will set non-convertible values to NaN

# Calculate the age of the company
current_year = 2023  # Assuming the current year is 2023
df['age'] = current_year - df['Founded']
```
